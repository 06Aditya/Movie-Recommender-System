```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
#Merging datasets on basis of title
movies = movies.merge(credits,on='title')
```

```python
movies.head(1)
```

```python
#imp columns for content based recommender system:
#genres
#id
#keywords
#title
#overview
#cast
#crew

movies = movies[['movie_id_x','title','overview','genres','keywords','cast_x','crew_x']]
```

```python
#Main Dataset
movies.head()
```

```python
#checking for missing values
movies.isnull().sum()
```

```python
#dropping missing values
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
#deleting duplicate values
movies.drop_duplicates().duplicated().sum()
```

```python
#1 element of genres to see the form of it.
movies.iloc[0].genres
```

```python
#'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
#['Action','Adventure','Fantasy','SciFi'] ->preprocessing to get genres in this form.
```

```python
import ast
#Helper Function
def convert(obj):
    L = []
    p = ast.literal_eval(str(obj))
    for i in p:
#     for i in obj:
        L.append(i["name"])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
import ast
#Helper Function for cast
def convert3(obj):
    L = []
    counter = 0
    p = ast.literal_eval(str(obj))
    for i in p:
        if counter !=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast_x'] = movies['cast_x'].apply(convert3)
```

```python
import ast
#Helper Function for crew
def fetchDirector(obj):
    L = []
    p = ast.literal_eval(str(obj))
    for i in p:
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew_x'] = movies['crew_x'].apply(fetchDirector)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
#converting string in overview in list
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#Removing space between words in genres,keywords,cast,crew
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast_x'] = movies['cast_x'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew_x'] = movies['crew_x'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast_x'] + movies['crew_x']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id_x','title','tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
#Stemming using NLTK
#['loved','loving','love']->['love','love','love']
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags']
```

```python
#TEXT VECTORIZATION using scikit learn
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000 ,stop_words='english' )
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names()
```

```python
#Cosine distance(angle between 2 angles)
#distance is indirectly proportional to similarity.
#Calculating cosine distance using scikit learn
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse = True,key = lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman')
```
